```
cerial_data = pd.read_csv("Cereals.csv")
print(cerial_data.columns)
print(cerial_data.loc[:,['calories', 'protein', 'fat', 'sodium', 'fiber',
       'carbo', 'sugars', 'potass', 'vitamins', 'shelf', 'weight', 'cups']])
```

```
cerial_ndata = cerial_data.loc[:,['calories', 'protein', 'fat', 'sodium', 'fiber','carbo', 'sugars', 'potass',  'vitamins', 'shelf', 'weight', 'cups', 'rating']]
#cerial_ndata = cerial_data.loc[:,['calories', 'protein', 'fat', 'sodium', 'rating']]
cerial_ndata
```

```
from sklearn.preprocessing import StandardScaler
c_std = StandardScaler().fit_transform(cerial_ndata)
```

```
c_std_df=pd.DataFrame(c_std)
c_std_df = c_std_df.dropna(axis=0, how='any')
c_std_df
```

```
import numpy as np
```

```
# features are columns from x_std
features = c_std_df.T 
covariance_matrix = np.cov(features)
print(covariance_matrix)
```

```
eig_vals, eig_vecs = np.linalg.eig(covariance_matrix)
print('Eigenvectors \n%s' %eig_vecs)
print('\nEigenvalues \n%s' %eig_vals)
```

```
# How much 1 eigenvector can explain the variances
eig_vals[0] / sum(eig_vals)
```

```
projected_X = c_std_df.dot(eig_vecs.T[0])
```

```
projected_X
```

```

from sklearn import decomposition
pca = decomposition.PCA(n_components=2)
sklearn_pca_x = pca.fit_transform(c_std_df)
pca.explained_variance_ratio_
```

```
pca_result = pd.DataFrame(sklearn_pca_x, columns=['PC1', 'PC2'])
pca_result
```
